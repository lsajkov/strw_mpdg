```python
import numpy as np
import matplotlib.pyplot as plt

from astropy.io import fits
from astropy.wcs import WCS
from astropy.nddata import Cutout2D

import astropy.units as u

plt.rcParams.update({
    'font.family': 'serif',
    'font.size': 24,
    'axes.labelsize': 'large',
    'mathtext.fontset': 'stix',
    'axes.linewidth': 1.5,
    'xtick.direction': 'in',
    'ytick.direction': 'in'
})
```

```python
with fits.open('/data2/lsajkov/mpdg/strw_mpdg/KiDS_DR4.0_227.0_-0.5_r_det_sci.fits') as hdu:
    header = hdu[0].header
    img = hdu[0].data
    wcs = WCS(hdu[0].header)
```

```python
cutout = Cutout2D(img, (210.5, -1.25), 2 * u.arcmin, wcs = wcs)
```
